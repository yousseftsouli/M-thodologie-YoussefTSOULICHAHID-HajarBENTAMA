```python
#These are the criterias that we want to match in our dataset : 
# -We are only interested in the videos published during the year 2021
# -We are only interested in videos that have more than 5000 likes
# -We are only interested in videos that gathered a total of more than 250 000 views during the year 2021
# -We are only interested in videos that have more than 150 dislikes

#This way, this scenario is more realistic and close to a real world problem and will help us filter out
#small and negligible videos
```

```python
#First thing we are going to do is split the publishet_at columns into columns to get the full date and the houre separately
df["Time"] = df["published_at"].str.split(" ", expand=True)[1]
df["Date"] = df["published_at"].str.split(" ", expand=True)[0]
```

```python
df
```

```python
#We are interested in the year so we are going to split the date columns into 3 columns, day - month and year
```

```python
df["Day"] = df["Date"].str.split("-", expand=True)[2]
df["Month"] = df["Date"].str.split("-", expand=True)[1]
df["Year"] = df["Date"].str.split("-", expand=True)[0]
```

```python
df
```

```python
#Before doing anything with the new columns we have to check their type and correct it if they are wrong.
df.dtypes
```

```python
#indeed, for the year, month and day, the type was wrong so we'll switch it from object to integer
df['Year'] = df['Year'].astype('int')
df['Month'] = df['Month'].astype('int')
df['Day'] = df['Day'].astype('int')
```

```python
df.dtypes
```

```python
#Now that the types are set, We delete all the rows that contain videos published during the year 2020
df = df[df['Year'] != 2020]
```

```python
import texthero as hero
from texthero import preprocessing
df_new['title'] = df_new['title'].pipe(hero.clean,[preprocessing.remove_urls , preprocessing.clean])
df_new['tags'] = df_new['tags'].pipe(hero.clean,[preprocessing.remove_urls , preprocessing.clean])
df_new
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bag_of_words_df = pd.DataFrame()

TOP_N_WORDS = 10

for text_column in ['title','tags']:
  vec = CountVectorizer(max_features = TOP_N_WORDS)
  txt_to_fts = vec.fit_transform(df_new[text_column]).toarray()
  txt_to_fts.shape

  names = vec.get_feature_names_out()
  print("Top repeated words for {}:".format(text_column),names)

  txt_fts_names = [text_column + f'_word_{i}_count' for i in range(TOP_N_WORDS)]
  bag_of_words_df[txt_fts_names] = txt_to_fts
```

```python
merged_df = pd.concat([df_new,bag_of_words_df],axis=1).drop(["tags","title"],axis=1)
```

```python
merged_df
```

```python
merged_df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
X = merged_df.drop(['dislikes'],axis=1)
y = merged_df['dislikes']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from catboost import CatBoostRegressor

folds = 5

regressor = CatBoostRegressor(silent=True)

pipeline = Pipeline([ ('estimator', regressor)])

cv = KFold(n_splits=folds)
scores = cross_val_score(pipeline, X, y,cv = cv , scoring='neg_mean_absolute_error')
```

```python
print(scores)
```

```python
scores.mean()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
cat_regressor = CatBoostRegressor(silent=True)
cat_regressor.fit(X_train,y_train)
```

```python
prediction_dislikes = cat_regressor.predict(X_test)
prediction_dislikes
```
