```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("youtube_dislike_dataset.csv")
```

```python
#We will proceed to the cleaning and preprocessing of our dataset
```

```python
#All the videos have been published in the years 2020 and 2021.
#Lets suppose that we are conducting a study for the year 2021, so we're only interested in the videos that were 
#published during 2021.
```

```python
df.count()
```

```python
#First thing we are going to do is split the publishet_at columns into columns to get the full date and the houre separately
df["Time"] = df["published_at"].str.split(" ", expand=True)[1]
df["Date"] = df["published_at"].str.split(" ", expand=True)[0]
```

```python
df
```

```python
#We are interested in the year so we are going to split the date columns into 3 columns, day - month and year
```

```python
df["Day"] = df["Date"].str.split("-", expand=True)[2]
df["Month"] = df["Date"].str.split("-", expand=True)[1]
df["Year"] = df["Date"].str.split("-", expand=True)[0]
```

```python
df
```

```python
#Before doing anything with the new columns we have to check their type and correct it if its wrong.
df.dtypes
```

```python
#indeed, the type was wrong so we'll switch it from object to integer
df['Year'] = df['Year'].astype('int')
df['Month'] = df['Month'].astype('int')
df['Day'] = df['Day'].astype('int')
```

```python
df.dtypes
```

```python
#We delete all the rows that contain videos published during the year 2020
df = df[df['Year'] != 2020]
```

```python
df
```

```python
df.count()
```

```python
#We check that we only have the year 2021 left as we want
df.Year.unique()
```

```python
#Now that our dataset contains only the information of the videos published in 2021, we have no more use for multiple columns
#such as the ids, the name of the channel as well the the date, day, month, year and hour of publishement of the videos
```

```python
df_new = df.drop(['video_id','published_at','channel_id','channel_title','Date','Day','Month','Year','Time'],axis=1)
```

```python
import texthero as hero
from texthero import preprocessing

custom_pipeline = [preprocessing.fillna,
                   preprocessing.remove_whitespace,
                   preprocessing.remove_urls,
                   ]

df_new['description'] = df_new['description'].pipe(hero.clean, custom_pipeline)
df_new['comments'] = df_new['comments'].pipe(hero.clean, custom_pipeline)
df_new['tags'] = df_new['tags'].pipe(hero.clean, custom_pipeline)
df_new
```

```python
pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
description_analysis = df_new['description'].apply(analyzer.polarity_scores).apply(pd.Series)
```

```python
tags_analysis = df_new['tags'].apply(analyzer.polarity_scores).apply(pd.Series)
```

```python
comments_analysis = df_new['comments'].apply(analyzer.polarity_scores).apply(pd.Series)
```

```python
description_analysis.rename(columns={"neg": "des_neg", "neu": "des_neu", "pos": "des_pos", "compound": "des_compound"},inplace=True)
comments_analysis.rename(columns={"neg": "com_neg", "neu": "com_neu", "pos": "com_pos", "compound": "com_compound"},inplace=True)
tags_analysis.rename(columns={"neg": "tag_neg", "neu": "tag_neu", "pos": "tag_pos", "compound": "tag_compound"},inplace=True)

sentiment_df = pd.concat([description_analysis,comments_analysis,tags_analysis],axis=1)
```

```python
cleaned_df = pd.concat([df_new,sentiment_df],axis=1).drop(["comments","description","tags","title"],axis=1)
```

```python
cleaned_df.columns
```

```python
X = cleaned_df.drop(['dislikes'],axis=1)
y = cleaned_df['dislikes']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from catboost import CatBoostRegressor
from xgboost import XGBRegressor

folds = 5

regressor = CatBoostRegressor(silent=True)

pipeline = Pipeline([ ('estimator', regressor)])

cv = KFold(n_splits=folds)
scores = cross_val_score(pipeline, X, y,cv = cv , scoring='neg_mean_absolute_error')
```
